# Life Expectancy vs. GDP
## How are a country's GDP and life expectancy related? 

Life expectancy data from the WHO
GDP data from the World Bank

```python
#Read in data & import modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv('all_data.csv')
```

```python
#Look at data
print(df.head(10))
countries = df.Country.unique()
print(countries)
years = df.Year.unique()
print(years)
```

```python
# Life Expectancy Summary Statistics
life_min = df['Life expectancy at birth (years)'].min()
life_max = df['Life expectancy at birth (years)'].max()
life_med = df['Life expectancy at birth (years)'].median()
life_avg = round(df['Life expectancy at birth (years)'].mean(),2)
print(f'The shortest life expectancy is {life_min} years')
print(f'The longest life expectancy is {life_max} years')
print(f'The average life expectancy is {life_avg} years, while the median is {life_med} years.')

plt.subplot(1,2,1)
plt.boxplot(df['Life expectancy at birth (years)'])
plt.title('Life Expectancy Boxplot')
plt.ylabel('Life expectancy at birth (years)')
plt.subplot(1,2,2)
plt.hist(df['Life expectancy at birth (years)'], bins=50, density=True)
plt.title('Life Expectancy Histogram')
plt.xlabel('Life expectancy at birth (years)')
```

### Life Expectancy Summary Statistics:
Life expectancies range from 44.3 years to 81.0 years, and are heavily skewed to the right with a median of 76.75 years. Life expectancies below 65 years are considered outliers. 

```python
# GDP Summary Statistics
gdp_min = df['GDP'].min()
gdp_max = df['GDP'].max()
gdp_med = df['GDP'].median()
gdp_q1 = np.quantile(df['GDP'], .25)
gdp_q3 = np.quantile(df['GDP'], .75)
gdp_avg = round(df['GDP'].mean(),2)
print(f'The lowest GDP is ${round(gdp_min/1000000000,2)} Billion')
print(f'The highest GDP is ${gdp_max/1000000000000} Trillion')
print(f'The average GDP is ${round(gdp_avg/1000000000000,2)} Trillion, while the median is ${round(gdp_med/1000000000000,2)} Trillion.')
print(f'50% of GDPs fall between ${round(gdp_q1/1000000000000,3)} Trillion and ${round(gdp_q3/1000000000000,2)} Trillion.')
print(f'GDPs above ${round((gdp_q3+(1.5*(gdp_q3-gdp_q1)))/1000000000000,3)} Trillion are considered outliers.')
plt.subplot(1,2,1)
plt.boxplot(df['GDP']/1000000000000)
plt.title('GDP Boxplot')
plt.ylabel('GDP (in Trillions)')
plt.subplot(1,2,2)
plt.hist(df['GDP']/1000000000000, bins=50)
plt.title('GDP Histogram')
plt.xlabel('GDP (in Trillions)')
```

### GDP Summary Statistics
- The lowest GDP is about 4.4B, and the higest GDP is 18B. 
- 25% of GDPs fall between 4.4B to 173B.
- 50% of GDPs fall between 173B and 4T.
- 25% of GDPs fall between 4T and 10T with a few outliers that go up to 18T.

```python
#GDP By Country Over Time
#Create empty dictionary with countries as keys
data_by_country = {}
for country in countries:
    data_by_country[country] = {'Years': [], 'Life_Expectancies': [], 'GDPs': []}

#Populate dictionary
for country in countries:
    for i in range(len(df.Country)):
        if df.Country[i]==country:
            data_by_country[country]['Years'].append(int(df.Year[i]))
            data_by_country[country]['Life_Expectancies'].append(float(df['Life expectancy at birth (years)'][i]))
            data_by_country[country]['GDPs'].append(int(df.GDP[i]))

#Plot
plt.figure(figsize=(12,12))
plt.subplot(2,1,1)
for country in countries:
    plt.plot(data_by_country[country]['Years'], data_by_country[country]['Life_Expectancies'], marker='s')
plt.legend(countries)
plt.subplot(2,1,2)
for country in countries:
    plt.plot(data_by_country[country]['Years'], data_by_country[country]['GDPs'], marker='s')
plt.legend(countries)
```

## GDP and Life Expectancies Over Time By Country:
### Life Expectancies:
- Zimbabwe has significancly lower life expectancy than any of the other 5 countries, however it has been steadily increasing since about 2006. In 2000, Zimbabwe's life expectancy was about 25 years less than the next highest (China). As of 2015, there is less than a 15 year gap.
- Germany has the highest life expectancy, followed by Chile, The USA, Mexico, and then China. All 5 of these countries have seen slight increases in life expectancy since 2000, and have remained within about 5 years of each other.
### GDPs:
- Zimbabwe, Chile, Mexico, and Germany have seen very little GDP growth from 2000 to 2015
- The US and China have seen significant GDP growth (about 750B and 850B respectively)
